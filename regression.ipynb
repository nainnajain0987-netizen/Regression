#Theory Questions

1. What is Simple Linear Regression
   - Simple Linear Regression (SLR) is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line: Y=mX+c
It predicts Y based on X.


2. What are the key assumptions of Simple Linear Regression
   - Linearity – The relationship between X and Y is linear.
Independence – Observations are independent.
Homoscedasticity – Constant variance of errors.
Normality of residuals – Errors are normally distributed.
No significant outliers


3. What does the coefficient m represent in the equation Y=mX+c
   - The slope m represents:
The change in Y for a one-unit increase in X.
Direction of relationship:
Positive → Y increases as X increases.
Negative → Y decreases as X increases.


4. What does the intercept c represent in the equation Y=mX+c
   - The intercept c is:
The predicted value of Y when X = 0.
It provides the baseline level of Y.

6. What is the purpose of the least squares method in Simple Linear Regression
   - The Least Squares Method finds the line that minimizes:
∑(y actual - y predicted)2
It ensures the best possible fit by minimizing squared errors


7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression
   - R² represents:
The proportion of variance in Y explained by X.
Ranges from 0 to 1.
0.8 → 80% of variation explained.


8. What is Multiple Linear Regression
   - Multiple Linear Regression (MLR) models the relationship between:
One dependent variable (Y)
Two or more independent variables (X₁, X₂, ..., Xₙ)
Y=β0​+β1​X1​+β2​X2​+...+βn​Xn​+ϵ

11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model
    - Heteroscedasticity = non-constant error variance.
Effects:
Biased standard errors
Invalid hypothesis tests
Less reliable confidence intervals


12. How can you improve a Multiple Linear Regression model with high multicollinearity
    - Remove correlated variables
Combine variables
Use Principal Component Analysis (PCA)
Use Ridge or Lasso regression
Center variables

    - Does not imply causation.
Always increases with more variables.
Doesn’t detect overfitting.
Doesn’t indicate bias.


18. How would you interpret a large standard error for a regression coefficient
    - A large standard error means:

Likely not statistically significant.
Possibly due to multicollinearity or small sample size

31. How is polynomial regression implemented in Python
    -

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Sample data
X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([1,4,9,16,25])

# Create polynomial regression model
degree = 2
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X, y)

# Predict
y_pred = model.predict(X)
```
